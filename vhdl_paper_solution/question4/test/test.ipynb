```python
import numpy as np
import matplotlib.pyplot as plt

periods = [16, 14, 12, 10, 8, 4, 2]

y_out = []

for period in periods:
    half_period = period // 2

    y_out.extend([1] * half_period)
    y_out.extend([0] * half_period)

y_out = np.array(y_out)
clock_cycles = np.arange(len(y_out))

print("Total clock cycles:", len(y_out))
print("Signal:", y_out)

plt.figure(figsize=(14, 4))
plt.step(clock_cycles, y_out, where="post")

plt.xlabel("Clock cycle")
plt.ylabel("y_out")
plt.yticks([0, 1])
plt.title("Chirp periods: 16, 14, 12, 10, 8, 4, 2 clocks")
plt.grid(True)

plt.show()
```
